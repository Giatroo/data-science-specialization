## Question 1
For this question, calculate the win/loss ratio's correlation with the population of the city it is in for the **NHL** using **2018** data.

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import re

def nhl_correlation(): 
    # reading DFs
    nhl_df=pd.read_csv("assets/nhl.csv")
    cities=pd.read_html("assets/wikipedia_data.html")[1]
    cities=cities.iloc[:-1,[0,3,5,6,7,8]]

    # Cleaning league DF
    double_name = {
        'New York' : 'NewYork',
        'San Jose' : 'SanFranciscoBay',
        'Los Angeles' : 'LosAngeles',
        'St. Louis' : 'StLouis',
        'New Orleans' : 'NewOrleans',
        'New Jersey' : 'NewYork',
        'Kansas City' : 'Kansas',
        'Salt Lake City' : 'SaltLake',
        'San Diego' : 'SanDiego',
        'San Antonio' :'SanAntonio',
        'Oklahoma City' : 'Oklahoma',
        'Tampa Bay' : 'TampaBay',
        'Golden State' : 'SanFranciscoBay',
        'San Francisco' : 'SanFranciscoBay',
        'New England' : 'Boston',
        'Green Bay' : 'GreenBay',
        'Arizona': 'Phoenix',
        'Colorado' :'Denver',
        'Texas': 'Dallas', 
        'Brooklyn' : 'NewYork',
        'Utah' : 'SaltLake',
        'Carolina' : 'Charlotte',
        'Minnesota' : 'Minneapolis',
        'Tennessee' : 'Nashville',
        'Oakland' : 'SanFranciscoBay',
        'Anaheim' : 'LosAngeles',
        'Florida' :'Miami',
        'Vegas' : 'LasVegas',
        'Indiana' : 'Indianapolis',
        'Indianapolispolis' : 'Indianapolis',
    }
    nhl_df = nhl_df[nhl_df['year'] == 2018]
    nhl_df = nhl_df.replace(regex='.*Division.*', value=np.nan).dropna()
    nhl_df['team'] = nhl_df['team'].str.replace('[*]', '')
    nhl_df = nhl_df[['team', 'W', 'L']]
    nhl_df['W'] = pd.to_numeric(nhl_df['W'])
    nhl_df['L'] = pd.to_numeric(nhl_df['L'])

    for key in double_name:
        nhl_df['team'] = nhl_df['team'].str.replace(f'^{key}', f'{double_name[key]}')

    nhl_df['city'] = nhl_df['team'].apply(str.split)
    nhl_df['team_name'] =  list(map(lambda x : x[1] if len(x) == 2 else x[1] + ' ' + x[2], list(nhl_df['city'])))
    nhl_df['city'] =  list(map(lambda x : x[0], list(nhl_df['city'])))

    # Cleaning cities DF
    cities_dict = {
        'City' : '',
        'Area' : '',
        '–.*' : '',
        ',.*' : '',
        '\s+' : ''
    }
    the_four = ['NFL', 'MLB', 'NBA', 'NHL']

    cities = cities.rename(columns={'Population (2016 est.)[8]':'Population'})
    cities = cities.applymap(str.strip)
    for league in the_four:
        cities[league] = cities[league].str.replace('\[.*\]','')
        cities[league] = cities[league].str.replace('[a-z](?P<letter>[A-Z])', lambda x : '|' + x.group('letter'))
        cities[league] = cities[league].replace('—', np.nan)
        cities[league] = cities[league].replace('', np.nan)

    # Little corner case =/
    cities.iloc[13,2] = np.nan

    for key in cities_dict:
        cities['Metropolitan area'] = cities['Metropolitan area'].str.replace(f'{key}', f'{cities_dict[key]}')

    # Preparing to merge the DFs
    teams_by_city = cities['NHL'].str.split('|')
    new_cities = pd.DataFrame(columns=cities.columns)
    for index in teams_by_city.index:
        if isinstance(teams_by_city[index], list):
            for team in teams_by_city[index]:
                my_row = cities.loc[index, :].copy()
                my_row['NHL'] = team
                new_cities = new_cities.append(my_row)
    cities = new_cities[['Metropolitan area', 'Population', 'NHL']].copy()

    team_indexed_nhl = nhl_df.set_index('team_name')
    team_indexed_cities = cities.set_index('NHL')
    team_indexed_cities = team_indexed_cities.rename(index={
        'Ranger' : 'Rangers',
        'Islander' : 'Islanders',
        'King' : 'Kings'
    })

    # Merging
    merged = team_indexed_cities.join(team_indexed_nhl, how='left')
    merged['ratio'] = merged['W'] / (merged['L']+merged['W'])
    merged['Population'] = pd.to_numeric(merged['Population'])

    # Getting the ratio and population values
    grouped = merged.groupby('Metropolitan area').agg(np.mean)[['Population', 'ratio']]
    population_by_region = list(grouped['Population'])
    win_loss_by_region = list(grouped['ratio'])
    assert len(population_by_region) == len(win_loss_by_region), "Q1: Your lists must be the same length"
    assert len(population_by_region) == 28, "Q1: There should be 28 teams being analysed for NHL"
    
    return stats.pearsonr(population_by_region, win_loss_by_region)[0]
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import re


def sports_team_performance():
    mlb_df=pd.read_csv("assets/mlb.csv")
    nhl_df=pd.read_csv("assets/nhl.csv")
    nba_df=pd.read_csv("assets/nba.csv")
    nfl_df=pd.read_csv("assets/nfl.csv")
    cities=pd.read_html("assets/wikipedia_data.html")[1]
    cities=cities.iloc[:-1,[0,3,5,6,7,8]]

    # Cleaning cities DF
    cities_dict = {
        'City' : '',
        'Area' : '',
        '–.*' : '',
        ',.*' : '',
        '\s+' : '',
        '[.]' : '',
    }
    the_four = ['NFL', 'MLB', 'NBA', 'NHL']

    cities = cities.rename(columns={'Population (2016 est.)[8]':'Population'})
    cities = cities.applymap(str.strip)
    for league in the_four:
        cities[league] = cities[league].str.replace('\[.*\]','')
        cities[league] = cities[league].str.replace('[a-z](?P<letter>[A-Z])', lambda x : '|' + x.group('letter'))
        cities[league] = cities[league].replace('—', np.nan)
        cities[league] = cities[league].replace('', np.nan)

    # Little corner case =/
    cities.iloc[13,2] = np.nan

    for key in cities_dict:
        cities['Metropolitan area'] = cities['Metropolitan area'].str.replace(f'{key}', f'{cities_dict[key]}')

    # Cleaning NFL DF
    double_name = {
        'New York' : 'NewYork',
        'San Jose' : 'SanFranciscoBay',
        'Los Angeles' : 'LosAngeles',
        'St. Louis' : 'StLouis',
        'New Orleans' : 'NewOrleans',
        'New Jersey' : 'NewYork',
        'Kansas City' : 'Kansas',
        'Salt Lake City' : 'SaltLake',
        'San Diego' : 'SanDiego',
        'San Antonio' :'SanAntonio',
        'Oklahoma City' : 'Oklahoma',
        'Tampa Bay' : 'TampaBay',
        'Golden State' : 'SanFranciscoBay',
        'San Francisco' : 'SanFranciscoBay',
        'New England' : 'Boston',
        'Green Bay' : 'GreenBay',
        'Arizona': 'Phoenix',
        'Colorado' :'Denver',
        'Texas': 'Dallas', 
        'Brooklyn' : 'NewYork',
        'Utah' : 'SaltLake',
        'Carolina' : 'Charlotte',
        'Minnesota' : 'Minneapolis',
        'Tennessee' : 'Nashville',
        'Oakland' : 'SanFranciscoBay',
        'Anaheim' : 'LosAngeles',
        'Florida' :'Miami',
        'Vegas' : 'LasVegas',
        'Indiana' : 'Indianapolis',
        'Indianapolispolis' : 'Indianapolis',
    }
    nfl_df = nfl_df[nfl_df['year'] == 2018]
    nfl_df = nfl_df.replace(regex='.*AFC.*', value=np.nan).dropna()
    nfl_df = nfl_df.replace(regex='.*NFC.*', value=np.nan).dropna()
    nfl_df['team'] = nfl_df['team'].str.replace('[*+]', '')
    nfl_df = nfl_df[['team', 'W', 'L']]
    nfl_df['W'] = pd.to_numeric(nfl_df['W'])
    nfl_df['L'] = pd.to_numeric(nfl_df['L'])

    for key in double_name:
        nfl_df['team'] = nfl_df['team'].str.replace(f'^{key}', f'{double_name[key]}')

    nfl_df['city'] = nfl_df['team'].apply(str.split)
    nfl_df['team_name'] =  list(map(lambda x : x[1] if len(x) == 2 else x[1] + ' ' + x[2], list(nfl_df['city'])))
    nfl_df['city'] =  list(map(lambda x : x[0], list(nfl_df['city'])))
    nfl_df['ratio nfl'] = nfl_df['W'] / (nfl_df['L']+nfl_df['W'])
    nfl_df = nfl_df[['team_name', 'city', 'ratio nfl']]
    group_nfl = nfl_df.groupby('city').agg(np.mean)

    mlb_df = mlb_df[mlb_df['year'] == 2018]
    mlb_df = mlb_df.replace(regex='.*Division.*', value=np.nan).dropna()
    mlb_df['team'] = mlb_df['team'].str.replace('[*]', '')
    mlb_df = mlb_df[['team', 'W', 'L']]
    mlb_df['W'] = pd.to_numeric(mlb_df['W'])
    mlb_df['L'] = pd.to_numeric(mlb_df['L'])

    for key in double_name:
        mlb_df['team'] = mlb_df['team'].str.replace(f'^{key}', f'{double_name[key]}')

    mlb_df['city'] = mlb_df['team'].apply(str.split)
    mlb_df['team_name'] =  list(map(lambda x : x[1] if len(x) == 2 else x[1] + ' ' + x[2], list(mlb_df['city'])))
    mlb_df['city'] =  list(map(lambda x : x[0], list(mlb_df['city'])))
    mlb_df['ratio mlb'] = mlb_df['W'] / (mlb_df['L']+mlb_df['W'])
    mlb_df = mlb_df[['team_name', 'city', 'ratio mlb']]
    group_mlb = mlb_df.groupby('city').agg(np.mean)

    nba_df = nba_df[nba_df['year'] == 2018]
    nba_df = nba_df.replace(regex='.*Division.*', value=np.nan).dropna()
    nba_df['team'] = nba_df['team'].str.replace('[*]', '')
    nba_df['team'] = nba_df['team'].str.replace('\(.+\)', '')
    nba_df = nba_df[['team', 'W', 'L']]
    nba_df['W'] = pd.to_numeric(nba_df['W'])
    nba_df['L'] = pd.to_numeric(nba_df['L'])

    for key in double_name:
        nba_df['team'] = nba_df['team'].str.replace(f'^{key}', f'{double_name[key]}')

    nba_df['city'] = nba_df['team'].apply(str.split)
    nba_df['team_name'] =  list(map(lambda x : x[1] if len(x) == 2 else x[1] + ' ' + x[2], list(nba_df['city'])))
    nba_df['city'] =  list(map(lambda x : x[0], list(nba_df['city'])))
    nba_df['ratio nba'] = nba_df['W'] / (nba_df['L']+nba_df['W'])
    nba_df = nba_df[['team_name', 'city', 'ratio nba']]
    group_nba = nba_df.groupby('city').agg(np.mean)

    nhl_df = nhl_df[nhl_df['year'] == 2018]
    nhl_df = nhl_df.replace(regex='.*Division.*', value=np.nan).dropna()
    nhl_df['team'] = nhl_df['team'].str.replace('[*]', '')
    nhl_df = nhl_df[['team', 'W', 'L']]
    nhl_df['W'] = pd.to_numeric(nhl_df['W'])
    nhl_df['L'] = pd.to_numeric(nhl_df['L'])

    for key in double_name:
        nhl_df['team'] = nhl_df['team'].str.replace(f'^{key}', f'{double_name[key]}')

    nhl_df['city'] = nhl_df['team'].apply(str.split)
    nhl_df['team_name'] =  list(map(lambda x : x[1] if len(x) == 2 else x[1] + ' ' + x[2], list(nhl_df['city'])))
    nhl_df['city'] =  list(map(lambda x : x[0], list(nhl_df['city'])))
    nhl_df['ratio nhl'] = nhl_df['W'] / (nhl_df['L']+nhl_df['W'])
    nhl_df = nhl_df[['team_name', 'city', 'ratio nhl']]
    nhl_df['city'] = nhl_df['city'].replace('Charlotte', 'Raleigh')
    group_nhl = nhl_df.groupby('city').agg(np.mean)


    sports = ['NFL', 'NBA', 'NHL', 'MLB']
    p_values = pd.DataFrame({k:np.nan for k in sports}, index=sports)

    nflXnba = group_nfl.join(group_nba, how='inner')
    nflXnhl = group_nfl.join(group_nhl, how='inner')
    nflXmlb = group_nfl.join(group_mlb, how='inner')
    nbaXnhl = group_nba.join(group_nhl, how='inner')
    nbaXmlb = group_nba.join(group_mlb, how='inner')
    nhlXmlb = group_nhl.join(group_mlb, how='inner')

    trash, p = stats.ttest_rel(nflXnba['ratio nfl'], nflXnba['ratio nba'])
    p_values.loc['NFL', 'NBA'] = p_values.loc['NBA', 'NFL'] = p
    trash, p = stats.ttest_rel(nflXnhl['ratio nfl'], nflXnhl['ratio nhl'])
    p_values.loc['NFL', 'NHL'] = p_values.loc['NHL', 'NFL'] = p
    trash, p = stats.ttest_rel(nflXmlb['ratio nfl'], nflXmlb['ratio mlb'])
    p_values.loc['NFL', 'MLB'] = p_values.loc['MLB', 'NFL'] = p
    trash, p = stats.ttest_rel(nbaXnhl['ratio nba'], nbaXnhl['ratio nhl'])
    p_values.loc['NBA', 'NHL'] = p_values.loc['NHL', 'NBA'] = p
    trash, p = stats.ttest_rel(nbaXmlb['ratio nba'], nbaXmlb['ratio mlb'])
    p_values.loc['NBA', 'MLB'] = p_values.loc['MLB', 'NBA'] = p
    trash, p = stats.ttest_rel(nhlXmlb['ratio nhl'], nhlXmlb['ratio mlb'])
    p_values.loc['NHL', 'MLB'] = p_values.loc['MLB', 'NHL'] = p

    assert abs(p_values.loc["NBA", "NHL"] - 0.02) <= 1e-2, "The NBA-NHL p-value should be around 0.02"
    assert abs(p_values.loc["MLB", "NFL"] - 0.80) <= 1e-2, "The MLB-NFL p-value should be around 0.80"
    return p_values
```
